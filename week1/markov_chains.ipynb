# Markov Chains Explained

A **Markov chain** is a mathematical system where the probability of transitioning to the next state depends **only on the current state**, not on the history of how the system arrived at that state. Think of it as a "**memoryless**" system, where the future is only influenced by the present.

## Breakdown with an Example

### States:
Imagine a weather model with two states:
- **Sunny**
- **Rainy**

### Transitions:
The weather can transition from:
- Sunny to Rainy
- Rainy to Sunny

But the probability of each transition is determined **only by the current weather state**, not the weather history.

### Example:
If **today is Sunny**:
- The probability of **tomorrow being Rainy** might be **30%**
- The probability of **tomorrow being Sunny** might be **70%**

The fact that it was sunny **yesterday** doesn't influence the probability of tomorrow being rainy or sunny.

### Markov Property:
The "**memoryless**" aspect is key. The probability of transitioning to the next state (**Rainy** or **Sunny**) depends **only on the current state** (**Sunny** or **Rainy**), not on any previous states.

### Simplified Analogy:
If you're in a room and can only move to one of two other rooms, the probability of moving to either of those rooms is determined **only by the room you're currently in**, not by the path you took to get there.

## Hidden Markov Models (HMM)

Example:
Markov Chain:
You’re modeling weather:

States: Sunny, Rainy

You see the actual weather each day.

You model transitions: P(Sunny→Rainy) = 0.3, etc.

Hidden Markov Model:
You want to infer someone's emotional state based on text:

Hidden States: Happy, Sad

Observations: Words like “yay!”, “ugh”

You don’t know their mood directly — you infer it from what they say.
