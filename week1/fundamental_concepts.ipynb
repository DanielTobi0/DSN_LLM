# Corpus

these represent a collection of data, which can be structured(tablular data) or unstructured(text, audio, video)

# Tokens

this is the basic units of text that a model processes.

they are the smallest elements the model 'understands' and works with during training and inference.

<br>

A token is typically:
* a character ('a', 'b', 'c')
* a word ("dog")
* a subword ("##ing" or "un##" from "tokenizing")

It depends on how the text is tokenized, which is the process of converting raw text into these discrete units.


[Hugging Face tokenizer](https://huggingface.co/learn/nlp-course/en/chapter2/4?fw=pt)

# Vocabulary

Vocabulary refers to the set of unique tokens (words or symbols) that appear in the corpus.
